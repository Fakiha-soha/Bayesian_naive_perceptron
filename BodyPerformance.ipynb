# Naive Bayesian Classifier

```python
from sklearn.model_selection import train_test_split
newdf = df2.copy()
x = newdf[['age',	'gender',	'height_cm',	'weight_kg','diastolic',	'systolic', 'class']]
y = newdf[['class']]

#splitting the dataset to 85:15 ratio for Train and Test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 0)
```

```python
def naive_bayeisian (x_train, age, gender,	height_cm,	weight_kg, diastolic,	systolic):
  #For class: 1 (D)

  # Calculate prior probability
  class_prior = x_train[x_train['class']==1].count()[0] / len(x_train)

  # Calculate the mean and standard deviation and likelihood 

  age_mean = x_train[x_train['class'] == 1]['age'].mean()
  age_std = x_train[x_train['class'] == 1]['age'].std()
  age_p = (1 / (np.sqrt(2 * np.pi) * age_std)) * np.exp(-((age - age_mean) ** 2 / (2 * age ** 2)))

  gender_p = x_train[( x_train['class'] == 1) ].count()[0] / x_train[x_train['class'] == 1].count()[0]

  height_cm_mean = x_train[x_train['class'] == 1]['height_cm'].mean()
  height_cm_std = x_train[x_train['class'] == 1]['height_cm'].std()
  height_cm_p = (1 / (np.sqrt(2 * np.pi) * height_cm_std)) * np.exp(-((height_cm - height_cm_mean) ** 2 / (2 * height_cm_std ** 2)))

  weight_kg_mean = x_train[x_train['class'] == 1]['weight_kg'].mean()
  weight_kg_std = x_train[x_train['class'] == 1]['weight_kg'].std()
  weight_kg_p = (1 / (np.sqrt(2 * np.pi) * weight_kg_std)) * np.exp(-((weight_kg - weight_kg_mean) ** 2 / (2 * weight_kg_std ** 2)))

  diastolic_mean = x_train[x_train['class'] == 1]['diastolic'].mean()
  diastolic_std = x_train[x_train['class'] == 1]['diastolic'].std()
  diastolic_p = (1 / (np.sqrt(2 * np.pi) * diastolic_std)) * np.exp(-((diastolic - diastolic_mean) ** 2 / (2 *diastolic_std ** 2)))

  systolic_mean = x_train[x_train['class'] == 1]['systolic'].mean()
  systolic_std = x_train[x_train['class'] == 1]['systolic'].std()
  systolic_p = (1 / (np.sqrt(2 * np.pi) *systolic_std)) * np.exp(-((systolic - systolic_mean) ** 2 / (2 *systolic_std ** 2)))

  #Posterior probability Class 1
  class_posterior = gender_p * class_prior * height_cm_p * weight_kg_p * diastolic_p * systolic_p 
  #print(class_posterior)

    #For class: 2 (C)
  # Calculate prior probability
  class_prior_2 = x_train[x_train['class'] == 2].count()[0] / len(x_train)

  # Calculate the mean and standard deviation and likelihood
  age_mean_2 = x_train[x_train['class'] == 2]['age'].mean()
  age_std_2 = x_train[x_train['class'] == 2]['age'].std()
  age_p_2 = (1 / (np.sqrt(2 * np.pi) * age_std_2)) * np.exp(-((age - age_mean_2) ** 2 / (2 * age_std_2 ** 2)))

  gender_p_2 = x_train[(x_train['class'] == 2)].count()[0] / x_train[x_train['class'] == 2].count()[0]
 
  height_cm_mean_2 = x_train[x_train['class'] == 2]['height_cm'].mean()
  height_cm_std_2 = x_train[x_train['class'] == 2]['height_cm'].std()
  height_cm_p_2 = (1 / (np.sqrt(2 * np.pi) * height_cm_std_2)) * np.exp(-((height_cm - height_cm_mean_2) ** 2 / (2 * height_cm_std_2 ** 2)))

  weight_kg_mean_2 = x_train[x_train['class'] == 2]['weight_kg'].mean()
  weight_kg_std_2 = x_train[x_train['class'] == 2]['weight_kg'].std()
  weight_kg_p_2 = (1 / (np.sqrt(2 * np.pi) * weight_kg_std_2)) * np.exp(-((weight_kg - weight_kg_mean_2) ** 2 / (2 * weight_kg_std_2 ** 2)))

  diastolic_mean_2 = x_train[x_train['class'] == 2]['diastolic'].mean()
  diastolic_std_2 = x_train[x_train['class'] == 2]['diastolic'].std()
  diastolic_p_2 = (1 / (np.sqrt(2 * np.pi) * diastolic_std_2)) * np.exp(-((diastolic - diastolic_mean_2) ** 2 / (2 * diastolic_std_2 ** 2)))

  systolic_mean_2 = x_train[x_train['class'] == 2]['systolic'].mean()
  systolic_std_2 = x_train[x_train['class'] == 2]['systolic'].std()
  systolic_p_2 = (1 / (np.sqrt(2 * np.pi) * systolic_std_2)) * np.exp(-((systolic - systolic_mean_2) ** 2 / (2 * systolic_std_2 ** 2)))

  # Posterior probability Class 2
  class_posterior_2 = gender_p_2 * class_prior_2 * height_cm_p_2 * weight_kg_p_2 * diastolic_p_2 * systolic_p_2
  #print(class_posterior_2)

  #For class: 3 (B)

  # Calculate prior probability
  class_prior_3 = x_train[x_train['class'] == 3].count()[0] / len(x_train)

  # Calculate the mean and standard deviation and likelihood
  age_mean_3 = x_train[x_train['class'] == 3]['age'].mean()
  age_std_3 = x_train[x_train['class'] == 3]['age'].std()
  age_p_3 = (1 / (np.sqrt(2 * np.pi) * age_std_3)) * np.exp(-((age - age_mean_3) ** 2 / (2 * age_std_3 ** 2)))

  gender_p_3 = x_train[(x_train['class'] == 3)].count()[0] / x_train[x_train['class'] == 3].count()[0]

  height_cm_mean_3 = x_train[x_train['class'] == 3]['height_cm'].mean()
  height_cm_std_3 = x_train[x_train['class'] == 3]['height_cm'].std()
  height_cm_p_3 = (1 / (np.sqrt(2 * np.pi) * height_cm_std_3)) * np.exp(-((height_cm - height_cm_mean_3) ** 2 / (2 * height_cm_std_3 ** 2)))

  weight_kg_mean_3 = x_train[x_train['class'] == 3]['weight_kg'].mean()
  weight_kg_std_3 = x_train[x_train['class'] == 3]['weight_kg'].std()
  weight_kg_p_3 = (1 / (np.sqrt(2 * np.pi) * weight_kg_std_3)) * np.exp(-((weight_kg - weight_kg_mean_3) ** 2 / (2 * weight_kg_std_3 ** 2)))

  diastolic_mean_3 = x_train[x_train['class'] == 3]['diastolic'].mean()
  diastolic_std_3 = x_train[x_train['class'] == 3]['diastolic'].std()
  diastolic_p_3 = (1 / (np.sqrt(2 * np.pi) * diastolic_std_3)) * np.exp(-((diastolic - diastolic_mean_3) ** 2 / (2 * diastolic_std_3 ** 2)))

  systolic_mean_3 = x_train[x_train['class'] == 3]['systolic'].mean()
  systolic_std_3 = x_train[x_train['class'] == 3]['systolic'].std()
  systolic_p_3 = (1 / (np.sqrt(2 * np.pi) * systolic_std_3)) * np.exp(-((systolic - systolic_mean_3) ** 2 / (2 * systolic_std_3 ** 2)))

  # Posterior probability Class 3
  class_posterior_3 = gender_p_3 * class_prior_3 * height_cm_p_3 * weight_kg_p_3 * diastolic_p_3 * systolic_p_3 
  #print(class_posterior_3)
  # For class: 4 (A)

  # Calculate prior probability
  class_prior_4 = x_train[x_train['class'] == 4].count()[0] / len(x_train)

  # Calculate the mean and standard deviation and likelihood
  age_mean_4 = x_train[x_train['class'] == 4]['age'].mean()
  age_std_4 = x_train[x_train['class'] == 4]['age'].std()
  age_p_4 = (1 / (np.sqrt(2 * np.pi) * age_std_4)) * np.exp(-((age - age_mean_4) ** 2 / (2 * age_std_4 ** 2)))

  gender_p_4 = x_train[(x_train['class'] == 4)].count()[0] / x_train[x_train['class'] == 4].count()[0]

  height_cm_mean_4 = x_train[x_train['class'] == 4]['height_cm'].mean()
  height_cm_std_4 = x_train[x_train['class'] == 4]['height_cm'].std()
  height_cm_p_4 = (1 / (np.sqrt(2 * np.pi) * height_cm_std_4)) * np.exp(-((height_cm - height_cm_mean_4) ** 2 / (2 * height_cm_std_4 ** 2)))

  weight_kg_mean_4 = x_train[x_train['class'] == 4]['weight_kg'].mean()
  weight_kg_std_4 = x_train[x_train['class'] == 4]['weight_kg'].std()
  weight_kg_p_4 = (1 / (np.sqrt(2 * np.pi) * weight_kg_std_4)) * np.exp(-((weight_kg - weight_kg_mean_4) ** 2 / (2 * weight_kg_std_4 ** 2)))

  diastolic_mean_4 = x_train[x_train['class'] == 4]['diastolic'].mean()
  diastolic_std_4 = x_train[x_train['class'] == 4]['diastolic'].std()
  diastolic_p_4 = (1 / (np.sqrt(2 * np.pi) * diastolic_std_4)) * np.exp(-((diastolic - diastolic_mean_4) ** 2 / (2 * diastolic_std_4 ** 2)))

  systolic_mean_4 = x_train[x_train['class'] == 4]['systolic'].mean()
  systolic_std_4 = x_train[x_train['class'] == 4]['systolic'].std()
  systolic_p_4 = (1 / (np.sqrt(2 * np.pi) * systolic_std_4)) * np.exp(-((systolic - systolic_mean_4) ** 2 / (2 * systolic_std_4 ** 2)))

  # Posterior probability
  class_posterior_4 = gender_p_4 * class_prior_4 * height_cm_p_4 * weight_kg_p_4 * diastolic_p_4 * systolic_p_4
  #print(class_posterior_4)
  
  if class_posterior > class_posterior_2 and class_posterior > class_posterior_3 and class_posterior > class_posterior_4:
      predicted_class = 1
  elif class_posterior_2 > class_posterior and class_posterior_2 > class_posterior_3 and class_posterior_2 > class_posterior_4:
      predicted_class = 2
  elif class_posterior_3 > class_posterior and class_posterior_3 > class_posterior_2 and class_posterior_3 > class_posterior_4:
    predicted_class = 3
  else:
    predicted_class = 4
  return predicted_class
```

#Perceptron

```python

#equation
def func(X, W):
  p= W[0]+ np.dot(X, W[1:])
  #print(X," ",p)
  return p

#activation function
def A(x):
  if x > 700:
      y = 4
  elif x > -1000 and x<0:
      y = 3
  elif x < -1000:
      y = 2
  else:
      y = 1
  return y

#classifier
def err(X, W):
  return A(func(X,W))
```

```python
def train(x_train, y_train):
  
  W = np.zeros(x_train.shape[1] + 1) #initializing weights
  eta = 0.05 #learning rate
  epoch = 0
  prev_mse = np.inf
  best_weights = None

  while True:
    class_pred_p = []
    for i in range(len(x_train)):
      X = x_train.iloc[i].values
      Ti = y_train[i][0]
      Pi = err(X, W)
      class_pred_p.append(Pi)
      delw = eta * (Ti - Pi)
      W[0] += delw * 1 #Update bias weight
      W[1:] += delw * X #Update feature weight
    
    MSE = np.mean((np.array(class_pred_p) - y_train) ** 2)

    epoch += 1

    print(f"Epoch {epoch}: MSE = {MSE}")
    if epoch > 1 and MSE >= prev_mse:
      break
    prev_mse = MSE

    if best_weights is None or MSE<best_mse:
      best_weights = np.copy(W)
      best_mse = MSE

  return W, MSE
```

```python
def test_accuracy(x_test, y_test, weights):
    correct_preds = 0
    for i in range(len(x_test)):
        X = x_test.iloc[i].values
        Ti = y_test[i][0]
        Pi = err(X, weights) 
        if Pi == Ti:
            correct_preds += 1
    
    accuracy = correct_preds / len(x_test)
    return accuracy
```

```python
x_train2 = pd.DataFrame(x_train) 
x_test2 = pd.DataFrame(x_test)  
y_train2 = np.array(y_train).reshape(-1, 1) #reshape to 2D Array
y_test2 = np.array(y_test).reshape(-1, 1) #reshape to 2D Array
```

# Accuracy

```python
w, train_mse = train(x_train2, y_train2)
accuracy_p = test_accuracy(x_test2, y_test2, w)
acc_p = accuracy_p * 100
print(f"Train MSE: {train_mse}")
print(f"Test Accuracy: {acc_p}%")
print(f"Learned Weights: {w}")
```
